```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline 
sns.set()
print('Setup Done!')
```

#**Data manipulation:**

```python
df = pd.read_csv('/content/data.csv')
df.head()
```

```python
df.shape
```

```python
data = df.iloc[:15]
data.head()
```

```python
data.shape
```

```python
# Delete every column that has at least one missing value : 
data.dropna(axis=1)
```

#**Heatmap:**

**Heatmaps help you visualize matrix-like data.**

```python
corr = data.corr()
sns.heatmap(corr)
```

#**Distribution Plot (a.k.a. Histogram):**

```python

sns.distplot(data.Vision)
```

#**Count Plot (a.k.a. Bar Plot):-**

```python
sub_df = data.iloc[:5]
```

```python

sns.countplot(x='Name', data=sub_df)
 
# Rotate x-labels
plt.xticks(rotation=-45)
```

#**Density Plot:-**

```python
sns.kdeplot(df.Overall, df.Potential)
```
